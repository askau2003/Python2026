### 2. Lav disse plots
* Du skal lave koden der generer disse plots. Kom så tæt på denne illustration som muligt. Når du er færdig kan du forsøge at forbedre plotenes udseende.


```python
import matplotlib.pyplot as plt
import numpy as np

# y = ax^2 + bx + c

a = 1
b = 0
c = 1

def parabel(x):
    return a*x**2 + b*x + c

# x-værdier
x = np.linspace(-5, 5, 400)  # 400 punkter fra -5 til 5

# y-værdier
y = parabel(x)

# Parabel
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=1)  # x-aksen
plt.axvline(0, color='black', linewidth=1)  # y-aksen
plt.grid(True)
plt.show()
```

```python
# Vektor
import matplotlib.pyplot as plt

# 0,0 = origo

import matplotlib.pyplot as plt

# Figur
plt.figure(figsize=(6,6))

# Rød vektor
plt.quiver(0, 0, -3, 2, color='r', angles='xy', scale_units='xy', scale=1, label='v1 = (-3,2)')

# Blå vektor
plt.quiver(0, 0, 5, 5, color='b', angles='xy', scale_units='xy', scale=1, label='v2 = (5,5)')

# Gitter og akser
plt.grid(True)
plt.axhline(0, color='black')  # x-akse
plt.axvline(0, color='black')  # y-akse
plt.xlim(-5,5)  # x-akse grænser
plt.ylim(-2,6)  # y-akse grænser

# Titel og legend
plt.title('Vectors')
plt.legend()
plt.show()
```
