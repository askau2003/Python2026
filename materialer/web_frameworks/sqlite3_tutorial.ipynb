# SQLite3 Tutorial Flow

This Jupyter Notebook demonstrates the flow of the SQLite3 tutorial from the Python documentation.

## Step 1: Import sqlite3 and Create a Connection

```python
import sqlite3

# Create a connection to the database (creates the file if it doesn't exist)
con = sqlite3.connect("tutorial.db")
```

## Step 2: Create a Cursor

```python
# Create a cursor to execute SQL statements
cur = con.cursor()
```

## Step 3: Create a Table

```python
# Create a table named 'movie' with columns for title, year, and score
cur.execute("CREATE TABLE movie(title, year, score)")
```

## Step 4: Verify Table Creation

```python
# Verify the table was created by querying sqlite_master
res = cur.execute("SELECT name FROM sqlite_master")
print(res.fetchone())  # Output: ('movie',)
```

## Step 5: Insert Data

```python
# Insert data into the 'movie' table
cur.execute("""
    INSERT INTO movie VALUES
        ('Monty Python and the Holy Grail', 1975, 8.2),
        ('And Now for Something Completely Different', 1971, 7.5)
""")

# Commit the transaction to save changes
con.commit()
```

## Step 6: Query Data

```python
# Query data from the 'movie' table
res = cur.execute("SELECT score FROM movie")
print(res.fetchall())  # Output: [(8.2,), (7.5,)]
```

## Step 7: Insert More Data Using executemany

```python
# Insert more data using executemany
data = [
    ("Monty Python Live at the Hollywood Bowl", 1982, 7.9),
    ("Monty Python's The Meaning of Life", 1983, 7.5),
    ("Monty Python's Life of Brian", 1979, 8.0),
]
cur.executemany("INSERT INTO movie VALUES(?, ?, ?)", data)
con.commit()  # Commit the transaction
```

## Step 8: Query and Iterate Over Results

```python
# Query and iterate over results
for row in cur.execute("SELECT year, title FROM movie ORDER BY year"):
    print(row)
```

## Step 9: Close and Reopen the Connection

```python
# Close the connection
con.close()

# Reopen the connection and verify data persistence
new_con = sqlite3.connect("tutorial.db")
new_cur = new_con.cursor()
res = new_cur.execute("SELECT title, year FROM movie ORDER BY score DESC")
title, year = res.fetchone()
print(f"The highest scoring Monty Python movie is {title!r}, released in {year}")

# Close the new connection
new_con.close()
```

## Summary

This notebook demonstrates the complete flow of the SQLite3 tutorial:
1. Import the `sqlite3` module.
2. Create a connection to the database.
3. Create a cursor to execute SQL statements.
4. Create a table.
5. Insert data into the table.
6. Query data from the table.
7. Use `executemany` to insert multiple rows.
8. Iterate over query results.
9. Close and reopen the connection to verify data persistence.

This flow provides a clear and concise overview of how to use the `sqlite3` module in Python.
